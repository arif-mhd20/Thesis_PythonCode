```python
def drop_non_numeric_columns(df):
    # Select non-numeric columns
    non_numeric_columns = df.select_dtypes(exclude=[np.number]).columns

    # Drop non-numeric columns from the dataframe
    df.drop(non_numeric_columns, axis=1, inplace=True)

    return df
```

```python
import tensorflow
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
X_train.shape
```

```python
# In[14]:


classifier = Sequential()
```

```python
# In[15]:


classifier.add(Dense(units=3, kernel_initializer='he_uniform', activation='relu', input_dim=25))
```

```python
# In[16]:


classifier.add(Dense(units = 3, kernel_initializer = 'he_uniform',activation='relu'))
```

```python
# In[17]:


classifier.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='sigmoid'))
```

```python
# In[18]:


classifier.compile(optimizer = 'Adamax', loss = 'categorical_crossentropy', metrics = ['accuracy'])

```

```python
# In[43]:


model_history = classifier.fit(X_train, y_train, validation_split=0.33, batch_size=10, epochs=3)
```

```python
# In[92]:


print(model_history.history.keys())
```

```python
# In[93]:


plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# In[77]:


plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
